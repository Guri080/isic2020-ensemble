```python
import pandas as pd
import matplotlib.pyplot as plt

paths = ["backbone_0_effnet_projector3res3eff.csv",
         "backbone_1_effnet_projector3res3eff.csv",
         "backbone_2_resnet_projector3res3eff.csv",
         "backbone_3_resnet_projector3res3eff.csv",
         "backbone_4_resnet_projector3res3eff.csv"]
root = "/home/gssodhi/snap/firmware-updater/224/Desktop/melanoma_detection/ensemble/data/projection"

# Create figure with 3 subplots
fig, axes = plt.subplots(len(paths), 2, figsize=(18, 4*len(paths)))

for i, path in enumerate(paths):
    df = pd.read_csv(f"{root}/{path}")
    
    # Plot loss
    df.plot(x='epoch', y=['train_loss', 'val_loss'], ax=axes[i, 0])
    
    # Plot accuracy
    df.plot(x='epoch', y=['train_acc', 'val_acc'], ax=axes[i, 1])
    
    # Plot f1 and auc
    
    axes[i, 0].set_ylabel(path.replace('.csv', ''))

plt.tight_layout()
plt.show()
```
